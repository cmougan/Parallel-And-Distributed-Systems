#### Q6- How would get the smallest size from each of the records? The result should provide a number for each line of the input

```python
%%bash

cut -f11 example.bed|head -10 | sed -e 's/,$//'|awk -F, '
{
    primero=$1;
    for(i=1;i<NF+1;i++)
        if($i<primero) 
            primero=$i;
    print primero
}'| head -5
```

#### Q7- How would you now sort the records so that the first number shown is the smallest exon size? Again, the answer must provide a sorted list of numbers for each line of the input.


```python
%%bash

cut -f11 example.bed| sed -e 's/,$//'|awk -F, '
{
    primero=$1;
    for(i=1;i<NF+1;i++)
        if($i<primero)
            primero=$i;
    print primero
}' | sort | head -5
```

#### Q8- Now Get the 10 largest exons of chr1 stored in example.bed.

```python
%%bash

grep 'chr1' example.bed|cut -f11| sed -e 's/,$//'|awk -F, '
{
    primero=$1;
    for(i=1;i<NF+1;i++)
        if($i<primero)
            primero=$i;
    print primero
}' | sort -nr | head -10
```

#### Q10- Get the first 10 records of jan2017articles.csv with largest number of comments from the original csv file

#### Q12 (1p): Now we are going to create a new articles.csv where we get a different output data layout using awk tool

INPUT: Post date,Content type,Author,Title,Comm count,Path,Tags,Word count


OUTPUT: Title;Comment count;Word count;Post date

For the next exercises we have build some exceptions do to some particular errors like line 4 and

```python
%%bash

awk -F, 'BEGIN { OFS = "," };
{
  if (NR == 1) print;
  else
  {
    field = 4; 
    while ($field !~ /^[0-9]+$/) {field++};
    
    i = 4;
    while (i < field) {printf $i; i++};
    
    printf OFS $(field) OFS;
    
    field_2 = field + 2; 
    while ($field_2 !~ /^[0-9]+$/) {field_2++};
    
    print $(field_2) OFS $1 
  } 
}' jan2017articles.csv > articles.csv 
```
